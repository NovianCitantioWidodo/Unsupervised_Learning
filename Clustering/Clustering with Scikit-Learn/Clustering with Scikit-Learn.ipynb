```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering, SpectralClustering

from luwiji.cluster import demo
```

## Sample Data

```python
X1, y1 = demo.blob_data()
X2, y2 = demo.moon_data()
X3, y3 = demo.circle_data()
```

## Visualize

```python
X = X1
plt.scatter(X[:, 0], X[:,1], s=10)
plt.axis("equal");
```

## K-Means clustering

```python
model = KMeans(n_clusters=4)
labels = model.fit_predict(X)
center = model.cluster_centers_

plt.figure(figsize=(6, 6))
plt.scatter(X[:, 0], X[:, 1], s=10, c=labels)
plt.scatter(center[:, 0], center[:, 1], c='w', edgecolors='k', s=50, linewidths=2)
plt.axis("equal")
```

## K-Means waekness
- Kesulitan untuk menentukan k <br>
    Solusi : elbow analysis, silhoutte, and others
- Hanya untuk data yang bisa dirata - ratakan <br>
    Solusi : K-modes, K-medoid, K-prototype
- Solusi tidak selalu optimal
- Hanya untuk data yang linear - separable

## Menentukan K (n_cluster)
### tanpa label -> Inertia / Elbow analysis

```python
X = X1
score = []

for k in range(2, 25):
    model = KMeans(n_clusters=k)
    labels = model.fit_predict(X)
    score.append(model.inertia_)
    
plt.figure(figsize=(10, 5))
plt.plot(range(2, 25), score, "r-")
plt.xlabel("n_cluster")
plt.ylabel("inertia")
```

### Silhoutte score

```python
from sklearn.metrics import silhouette_score
```

```python
X = X1
score = []

for k in range(2, 25):
    model = KMeans(n_clusters=k)
    labels = model.fit_predict(X)
    siluet = silhouette_score(X, labels, metric='euclidean')
    score.append(siluet)
    
plt.figure(figsize=(10, 5))
plt.plot(range(2, 25), score, "r-")
plt.xlabel("silhoutte")
plt.ylabel("inertia")
```
